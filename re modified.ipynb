```python
import librosa
import soundfile as sf
import numpy as np
import scipy.signal as sig
from matplotlib import pyplot as plt
from scipy.io import wavfile
import os
```

```python
def voc2voc(input_path, output_path):
    # Load the audio file
    audio, sr = librosa.load(input_path, sr=None)

    # Adjust the pitch to make it sound deeper
    n_steps = -0.2
    audio_deeper = librosa.effects.pitch_shift(y=audio, sr= sr, n_steps=n_steps,bins_per_octave=2)

    b, a = sig.iirfilter(4, Wn=9000, fs=sr, btype="low", ftype="butter")
    y_lfilter = sig.lfilter(b, a, audio_deeper)

    augmented_audio = librosa.effects.time_stretch(y_lfilter, rate = 1)
    
    # Save the modified audio to a new file
    sf.write(output_path, augmented_audio, sr)
```

```python
input_folder = "selected_clip_wav"
output_folder = "mod_wav2"
# Iterate through each file in the input folder
for filename in os.listdir(input_folder): #loop through fie in folder
    file_path = os.path.join(input_folder,filename) #get file directory in folder
    file_path_output = os.path.join(output_folder,filename)
    voc2voc(file_path,output_folder)
```

```python
folder_path = 'selected_clip_wav' # Replace with the actual folder path
output = 'modified_wave'
# Iterate through each directory and file in the folder
for root, dirs, files in os.walk(folder_path):
    for file_name in files:
        file_path = os.path.join(root, file_name)
        output_path = os.path.join(output, file_name)
        voc2voc(file_path,output_path)
        print('modified ', file_name, ' finished')
```

```python
import concurrent.futures

folder_path = 'selected_clip_wav' # Replace with the actual folder path
output = 'mod_wav2'

def is_file_in_directory(filename, directory):
    file_path = os.path.join(directory, filename)
    return os.path.isfile(file_path)

def track_progress(futures, total_count, progress_callback):
    count = 0

    for future in concurrent.futures.as_completed(futures):
        # Process the completed future
        result = future.result()

        # Update the progress count
        count += 1
        progress_callback(count, total_count)
        
def print_progress(count, total_count):
    print(f"Processed {count}/{total_count} files.")

# Create a thread pool with a specified number of worker threads
with concurrent.futures.ThreadPoolExecutor(max_workers=20) as executor:
    # Submit tasks to the thread pool
    for root, dirs, files in os.walk(folder_path):
        for file_name in files:
            if not is_file_in_directory(file_name, output):
                file_path = os.path.join(root, file_name)
                output_path = os.path.join(output, file_name)          
                futures = [executor.submit(voc2voc, file_path,output_path ) ]

    track_progress(futures, 1, print_progress)
```
