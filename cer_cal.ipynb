```python
import pandas as pd

df = pd.read_csv("transcribe_trans.csv")

df['CER'] = None
df
```

```python
import re

def nom_text(txt):
    nom_txt = txt.replace(" ", "")
    nom_txt = nom_txt.translate ({ord(c): "" for c in "!@#$%^&*()[]{};:,./<>?\|`~-=_+"})
    return nom_txt
```

```python
from tqdm.auto import tqdm

for i in tqdm(range(len(df['path']))):
    try:
        df['sentence'][i] = nom_text(df['sentence'][i])
        df['transcribe'][i] = nom_text(df['transcribe'][i])
    except Exception as e:
        print("error occur:" , e, "at " ,i)
        print("sentence:",df['sentence'][i],"trans",df['transcribe'][i])
```

```python
df = df[df['transcribe'].notna()]
```

```python
df.to_csv('transcribe_trans_clean.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv("transcribe_trans_clean.csv")

df_filtered = df[df['CER'] <= 50]
df_filtered.to_csv('transcribe_trans_clean_limit_CER.csv', index=False)
```

```python
import shutil
import os
from tqdm.auto import tqdm

column_list = df_filtered['path'].tolist()
# Source folder path
source_folder = 'selected_clip_wav'

# Destination folder path
destination_folder = 'clean_voice'

# Iterate through the list of filenames
for filename in tqdm(column_list):
    source_file = os.path.join(source_folder, filename)
    if os.path.isfile(source_file):
        destination_file = os.path.join(destination_folder, filename)
        shutil.copy(source_file, destination_file)
    else:
        print(f"File '{filename}' not found in the source folder.")
```

```python
import librosa
folder_path = 'clean_voice'
total_duration_sec = 0.0

for filename in tqdm(os.listdir(folder_path)): #loop through fie in folder
    file_path = os.path.join(folder_path,filename) #get file directory in folder

    audio, sr = librosa.load(file_path, sr=None) #get audio data and sample rte
    duration = librosa.get_duration(y=audio,sr=sr) #get duration in sec
    total_duration_sec += duration #accumulate duration in sec

print('total duration:',total_duration_sec/3600, "h")

```
